```python
from math import ceil, floor
def float_round(num, places = 0, direction = floor):
    return direction(num * (10**places)) / float(10**places)
```

## Homer mushroom individual proteins

```python
for protein in proteins:

    dfmeans=pd.read_csv(expdir[:-10]+'/Output/Correlation - protein intensity to homer intensity/'+protein+'_homer_dfmeans_mushroom.csv')

    sns.set(font_scale=2.5)
    sns.set(font="Arial")
    sns.set_style("ticks")
    sns.despine()
    sns.set_palette("Set2")
    var="mean protein"

    df2plot=dfmeans

    fig=plt.figure(figsize=(5,5))
    g=sns.scatterplot(x="mean homer int in bin overall",y="mean protein",
                    data=df2plot,s=80,color="#2F67F8",linewidth=1,edgecolor='grey')

    plt.errorbar(x = df2plot["mean homer int in bin overall"].unique()
    ,
                 y=[df2plot[df2plot["bin number"]==0][var].mean(),
                    df2plot[df2plot["bin number"]==1][var].mean(),
                    df2plot[df2plot["bin number"]==2][var].mean(),
                    df2plot[df2plot["bin number"]==3][var].mean(),
                    df2plot[df2plot["bin number"]==4][var].mean()],
                 yerr=[df2plot[df2plot["bin number"]==0][var].sem(),
                       df2plot[df2plot["bin number"]==1][var].sem(),
                       df2plot[df2plot["bin number"]==2][var].sem(),
                       df2plot[df2plot["bin number"]==3][var].sem(),
                       df2plot[df2plot["bin number"]==4][var].sem()],
                 xerr=None,ecolor='black',fmt='none',capsize=10,linewidth=3,capthick=3,alpha=.7,)

    plt.errorbar(x = df2plot["mean homer int in bin overall"].unique()
    ,
                 y=[df2plot[df2plot["bin number"]==0][var].mean(),
                    df2plot[df2plot["bin number"]==1][var].mean(),
                    df2plot[df2plot["bin number"]==2][var].mean(),
                    df2plot[df2plot["bin number"]==3][var].mean(),
                    df2plot[df2plot["bin number"]==4][var].mean()],
                 yerr=None,
                 xerr=[.14,.14,.14,.14,.14],ecolor='black',fmt='none',capsize=0,linewidth=3,capthick=3,alpha=.7,)


    plt.rcParams['font.sans-serif'] = 'Helvetica'
    g.grid(False)
    #plt.ylim([0,1.5])
    #plt.xlim([.2,2.8])

    #fit linear curve
    dfnonans=df2plot.dropna()
    x=dfnonans["mean homer int in bin overall"].unique()
    y=dfnonans.groupby("mean homer int in bin overall").mean()["mean protein"].values
    m,b = np.polyfit(x, y, 1) 
    plt.plot( x, m*x+b, 'k',linestyle='dashed',alpha=.4,linewidth=3) 
    
    from pydoc import help
    from scipy.stats.stats import pearsonr, spearmanr
    [pearson, pval]=pearsonr(dfnonans["mean protein"],dfnonans["mean homer int in bin overall"])
    [spearman, pval2]=spearmanr(dfnonans["mean protein"],dfnonans["mean homer int in bin overall"])
    
    pearsonr=float_round(pearson**2,3)
    pval=float_round(pval,8)
    spearmanr=float_round(spearman**2,3)
    pval2=float_round(pval2,8)
    #print(str(pearson**2))
    #print(pval)

    #formatting
    g.xaxis.set_tick_params(width=3)
    g.yaxis.set_tick_params(width=3)
    for axis in ['top','bottom','left','right']:
      g.spines[axis].set_linewidth(4)
      g.spines[axis].set_color('#333333')
      g.spines[axis].set_alpha(1)
    offsets={"left":10,"bottom":0}
    trims={"left":True,"bottom":True,"right":False}
    sns.despine(ax=g,right=True,top=True,bottom=False,trim=trims,offset=offsets)
    g.set_ylabel(protein,color='#333333',size=20)
    g.set_xlabel('Homer',color='#333333',size=20)
    plt.yticks(g.get_yticks(),color='#333333',size=20)
    plt.xticks(g.get_xticks(),color='#333333',size=20,rotation=45)
    plt.title('Mushroom \n R^2 = '+str(pearsonr)+', p = '+str(pval)+'\n Spearman Rho: '+str(spearmanr)
             +', p = '+str(pval2))

    
    fig.savefig(expdir[:-10]+'/Output/Correlation - protein intensity to homer intensity/mushroom_'+protein+'.eps', bbox_inches='tight',format='eps')

    print('homer mushroom: '+str(len))
```

```python
for protein in proteins:

    dfmeans=pd.read_csv(expdir[:-10]+'/Output/Correlation - mean protein spot size to spine size (no. pixels phalloidin mask)/'+protein+'_dfmeans_mushroom.csv')
    dfmeans["mean size int in bin overall"]*=0.0009
    dfmeans["mean protein"]*=0.0009
    
    sns.set(font_scale=2.5)
    sns.set(font="Arial")
    sns.set_style("ticks")
    sns.despine()
    sns.set_palette("Set2")
    var="mean protein"

    df2plot=dfmeans

    fig=plt.figure(figsize=(5,5))
    g=sns.scatterplot(x="mean size int in bin overall",y="mean protein",
                    data=df2plot,s=80,color="#2F67F8",linewidth=1,edgecolor='grey')

    plt.errorbar(x = df2plot["mean size int in bin overall"].unique()
    ,
                 y=[df2plot[df2plot["bin number"]==0][var].mean(),
                    df2plot[df2plot["bin number"]==1][var].mean(),
                    df2plot[df2plot["bin number"]==2][var].mean(),
                    df2plot[df2plot["bin number"]==3][var].mean(),
                    df2plot[df2plot["bin number"]==4][var].mean()],
                 yerr=[df2plot[df2plot["bin number"]==0][var].sem(),
                       df2plot[df2plot["bin number"]==1][var].sem(),
                       df2plot[df2plot["bin number"]==2][var].sem(),
                       df2plot[df2plot["bin number"]==3][var].sem(),
                       df2plot[df2plot["bin number"]==4][var].sem()],
                 xerr=None,ecolor='black',fmt='none',capsize=10,linewidth=3,capthick=3,alpha=.7,)

    plt.errorbar(x = df2plot["mean size int in bin overall"].unique()
    ,
                 y=[df2plot[df2plot["bin number"]==0][var].mean(),
                    df2plot[df2plot["bin number"]==1][var].mean(),
                    df2plot[df2plot["bin number"]==2][var].mean(),
                    df2plot[df2plot["bin number"]==3][var].mean(),
                    df2plot[df2plot["bin number"]==4][var].mean()],
                 yerr=None,
                 xerr=[.03,.03,.03,.03,.03],ecolor='black',fmt='none',capsize=0,linewidth=3,capthick=3,alpha=.7,)


    plt.rcParams['font.sans-serif'] = 'Helvetica'
    g.grid(False)
    #plt.ylim([0,1.5])
    #plt.xlim([.2,2.8])

    #fit linear curve
    dfnonans=df2plot.dropna()
    x=dfnonans["mean size int in bin overall"].unique()
    y=dfnonans.groupby("mean size int in bin overall").mean()["mean protein"].values
    m,b = np.polyfit(x, y, 1) 
    plt.plot( x, m*x+b, 'k',linestyle='dashed',alpha=.4,linewidth=3) 
    
    from pydoc import help
    from scipy.stats.stats import pearsonr, spearmanr
    [pearson, pval]=pearsonr(dfnonans["mean protein"],dfnonans["mean size int in bin overall"])
    [spearman, pval2]=spearmanr(dfnonans["mean protein"],dfnonans["mean size int in bin overall"])
    
    pearsonr=float_round(pearson**2,3)
    pval=float_round(pval,8)
    spearmanr=float_round(spearman**2,3)
    pval2=float_round(pval2,8)
    #print(str(pearson**2))
    #print(pval)

    #formatting
    g.xaxis.set_tick_params(width=3)
    g.yaxis.set_tick_params(width=3)
    for axis in ['top','bottom','left','right']:
      g.spines[axis].set_linewidth(4)
      g.spines[axis].set_color('#333333')
      g.spines[axis].set_alpha(1)
    offsets={"left":10,"bottom":0}
    trims={"left":True,"bottom":True,"right":False}
    sns.despine(ax=g,right=True,top=True,bottom=False,trim=trims,offset=offsets)
    g.set_ylabel(protein+'\nProtein spot size (microns sq)',color='#333333',size=20)
    g.set_xlabel('Size (microns sq)',color='#333333',size=20)
    plt.yticks(g.get_yticks(),color='#333333',size=20)
    plt.xticks(g.get_xticks(),color='#333333',size=20,rotation=45)
    plt.title('Mushroom \n R^2 = '+str(pearsonr)+', p = '+str(pval)+'\n Spearman Rho: '+str(spearmanr)
             +', p = '+str(pval2))

    
    fig.savefig(expdir[:-10]+'/Output/Correlation - mean protein spot size to spine size (no. pixels phalloidin mask)/mushroom_'+protein+'.eps', bbox_inches='tight',format='eps')

    print('size mushroom: '+str(len))
```
