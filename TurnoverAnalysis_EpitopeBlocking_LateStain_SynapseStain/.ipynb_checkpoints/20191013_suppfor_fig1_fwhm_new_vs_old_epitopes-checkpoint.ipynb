## read second csv, plot bar plots

```python
df_barplots=pd.read_csv('/Volumes/ukps-all$/user/tdanko/Imaging/1_ECM/Fixed culture/STEDBIN/201909_silvio_analysis_fwhm_new_vs_old_epitopes/tenascin_aligned/usethisfile_data4.CSV',
                        header=None,sep=',')
```

```python
df_secondpeak={"New or old":["New epitopes","New epitopes","New epitopes","All epitopes",
                            "All epitopes","All epitopes"],
              "FWHM":df_barplots[3].tolist()+df_barplots[4].tolist(),
               "percent of first peak":df_barplots[0].tolist()+df_barplots[1].tolist()}
df_secondpeak=pd.DataFrame(df_secondpeak)
df_secondpeak["FWHM nanometes"]=df_secondpeak["FWHM"]*20
```

## plot2: second peak fwhm

```python
sns.set(font_scale=2.5)
sns.set_style("ticks")
sns.set_palette("Set2")
from matplotlib import ticker


var="FWHM nanometes"
fig=plt.figure(figsize=(4,6))

order=["New epitopes","All epitopes"]

g=sns.swarmplot(data=df_secondpeak,y=var,x="New or old",color="#2F67F8",order=order,size=10)

g=sns.barplot(data=df_secondpeak,y=var,x="New or old"
            ,order=order,color="#abc2fc",ci=None)

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value
        # we change the bar width
        patch.set_width(new_value)
        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(g, .7)

plt.errorbar(x=np.arange(0,len(order)),
             y=[df_secondpeak[df_secondpeak["New or old"]==condtype][var].mean() for condtype in order],
             yerr=[df_secondpeak[df_secondpeak["New or old"]==condtype][var].sem() for condtype in order],
             xerr=None,ecolor='black',fmt='none',capsize=10,linewidth=3,
             capthick=3,zorder=10,
            alpha=1)

'''plt.errorbar(x=np.arange(0,len(order)),
             y=[df_secondpeak[df_secondpeak["New or old"]==condtype][var].mean() for condtype in order],
             yerr=None,
             xerr=np.repeat(0.5, 5),ecolor='black',fmt='none',capsize=0,linewidth=3,
             capthick=30,zorder=10,
            alpha=1)'''
plt.axhline(0, color='#333333',alpha=1,linewidth=10)

#plt.axhline(negctrlmean, color='#333333',alpha=1,linewidth=5,linestyle='--')



g.xaxis.set_tick_params(width=4)
g.yaxis.set_tick_params(width=4)

g.set_ylabel('')
g.set_xlabel('',color='#333333',alpha=.7)
plt.xticks(g.get_xticks(),order,
           rotation=90,size=20)
#plt.yticks([0,20,40,60,80,100,120],size=20,color='#333333')
g.grid(False)
plt.ylim([0,600])
#plt.xlim([-1,len(order)])

for axis in ['top','bottom','left','right']:
  g.spines[axis].set_linewidth(4)
  g.spines[axis].set_color('#333333')
  g.spines[axis].set_alpha(1)



x1, x2 = 0,1
y= 590
h=5
col='black'
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=3.5, c=col)
plt.text((x1+x2)*.5, y+h, "*", ha='center', va='bottom', color=col,size=30, weight='bold')




offsets={"left":10,"bottom":0}
trims={"left":True,"bottom":False,"right":False}
sns.despine(ax=g,right=True,top=True,bottom=True,trim=trims,offset=offsets)

plt.xticks(g.get_xticks(),['New epitopes','All epitopes'], rotation=45,size=24)

diri="/Volumes/Macintosh HD/Users/taldankovich/Downloads/figures/"

fig.savefig(diri+'supp1_fwhm_secondpeak.svg', bbox_inches='tight', format='svg')




```

```python
#ttest
from scipy import stats
new=df_secondpeak[df_secondpeak["New or old"]=="New epitopes"][var]#.values()
old=df_secondpeak[df_secondpeak["New or old"]=="All epitopes"][var]#.values()

[stat,pval]=stats.ttest_ind(new,old)
print(pval)
```
